<center><h1>Численное интегрирование</h1></center>
Рассматривается задача о вычислении интеграла с помощью конечного числа значений интегрируемой функции.    
Пусть $ [a,b] $ - некоторый конечный отрезок и необходимо вычислить интеграл $ \int\limits_a^b F(x)dx $. Предположим что мы ставим задачу найти его приближенное значение по $ n $ значениям $ F(x_i) $ функции $ F $ в точках 
$$ x_i \in [a,b], i=1,2,...,n $$
Многие квадратурные методы основаны на замене функции $ F(x) $ на отрезке $ [a,b] $ или на части этого отрезка на более простую функцию $ \phi(x) $, например алгебраический многочлен. Если же сама функция $ F(x) $ или ее производные невысокого порядка имеют особенности или даже обращаются в бесконечность, то это затрудняет приближение или даже делает его невозможным. В этом случае необходимо избавиться от таких особенностей путем их выделения с помощью представления $ F(x) = p(x)f(x) $, где $ p(x) $ имеет такие же особенности, а $ f(x) $ достаточно гладкая функция. Функция $ p(x) $ называется весовой.  
Будем строить формулы имеющие вид 
$$ \int\limits_a^b p(x)f(x)dx \approx \sum\limits_{k=0}^{n} A_k f(x_k) $$

## Интерполяционные квадратурные формулы
Приведенные ИКФ основаны на приближении функции алгебраическими многочленами   
Квадратурная формула имеет алгебраическую степень точности (АСТ) $ m $, если она точна для всех многочленов степени $ \leq m$ и неточна хотя бы для одного многочлена степени $ m+1 $  
При интегрировании часто берут равноотстоящие узлы, интерполяционные формулы, построенные с такими узлами принято называть <b><i> формулами Ньютона-Котеса </i></b>. Их общий вид 
$$ I(f) = (b-a)\sum_{k=0}^{n}B_k^n f(a+kh) $$  
$$ h = \frac{b-a}{n},     B_k^n = \frac{-1^{n-k}}{nk!(n-k)!} \int_0^np(a+th)\frac{t(t-1)...(t-n)}{t-k}dt, p=1 $$  

| Формула                 | $I(f)$               | $R(f)$                 | ACT        | $I_c(f)$        | $R_c(f) $     |
| ----------------------- | -------------------- |----------------------- | ---------- | --------------- |---------------|
| Левые прямоугольники    | $ (b-a)f(a) $ | $ \frac{(b-a)^2}{2}f'(\eta) $| 0 | $ h\sum_{k=0}^{N-1}f_{k} $ |$ \frac{(b-a)^2}{2N}f'(\eta) $|
| Правые прямоугольники    | $ (b-a)f(b) $ | $ -\frac{(b-a)^2}{2}f'(\eta) $| 0 | $ h\sum_{k=0}^{N-1}f_{k+1} $ |$ -\frac{(b-a)^2}{2N}f'(\eta) $|
| Cредние прямоугольники    | $ (b-a)f(\frac{b+a}{2}) $ | $ \frac{(b-a)^3}{24}f''(\eta) $| 1 | $  h\sum_{k=0}^{N-1} f(a+kh+\frac{h}{2}) $ |$ \frac{(b-a)^3}{24N^2}f''(\eta) $|
|  Метод трапеций  |$ \frac{b-a}{2}(f(a) + f(b)) $ | $ -\frac{(b-a)^3}{12}f''(\eta) $| 1 | $  h(\frac{f_0 +f_N}{2} + \sum_{k=0}^{N-1}f_k) $ |$ -\frac{(b-a)^3}{12N^2}f''(\eta) $|
| Метод Симпсона    | $ \frac{b-a}{6}(f(a) + 4f(\frac{a+b}{2}) + f(b)) $ | $ \frac{(b-a)^5}{2880}f^{(4)}(\eta) $| 3 | $  \frac {h}{3}[f_0 + 4f_1 +2f_2 + 4f_3 + ... + f_N] $ |$ \frac{(b-a)^5}{180N^4}f^{(4)}(\eta) $|

Здесь $ N $ - число отрезков равной длины на которые разбивается исходный отрезок интегрирования, $ h = \frac{b-a}{N} $, $ f_k = f(a+kh) $

## Правило Рунге
Обозначим точное значение интеграла $ I = \int\limits_a^b f(x)dx $, а его приближенное значение полученное по некоторой квадратурной формуле порядка точности $ p $ с шагом $ h_k $ через $ I_k $. Тогда имеет место соотношение $ I = I_k + \alpha h_k^p $.  
Вычислим значения интеграла по указанной формуле с шагами $ h_1, h_2 $ и получим соотношение
$$ \begin{equation*}
       \begin{cases}
            I = I_1 + \alpha h_k^p \\
            I = I_2 + \alpha h_2^p \\
       \end{cases}
   \end{equation*}
$$  

$$ I_2 - I_1 = \alpha (h_1^p - h_2^p) $$
$$ \alpha = \frac{I_2 - I_1}{h_1^p - h_2^p} = \frac{I_2 - I_1}{ h_1^p(1 - (\frac{h_2}{h_1})^p )} $$
Таким образом для достижения точности $ \varepsilon $ при вычислении интеграла необходимо, чтобы выполнялось соотношение 
$$ | \frac{I_2 - I_1}{ h_1^p(1 - (\frac{h_2}{h_1})^p ) h_1^p} | = |\frac{I_2 - I_1}{1 - (\frac{h_2}{h_1})^p} | \leq \varepsilon $$

| Формула                 | Порядок точности($ p $)  |
| ----------------------- | -------------------- |
| Левые и правые прямоугольники | 1 |
| Средние прямоугольники и трапеций | 2 |
| Симпсона | 4 |

## Квадратурные формулы НАСТ Гаусса
Наивысшая алгебраическая степень точности (НАСТ), которой можно достичь при построении квадратурной формулы по $ n $ узлам - $ 2n -1 $.
Пусть необходимо вычислить $ \int\limits_{-1}^{1} f(x)dx $. Для построения квадратурной формулы НАСТ Гаусса в качестве узлов необходимо выбрать корни многочлена Лежандра 
$$ P_n(x) = \frac{1}{2^n n!}\frac{d^n}{dx^n}(x^2 - 1)^n $$
А коэффициенты могут быть вычислены следующим образом 
$$ A_k = \frac{2(1-x_k^2)}{n^2P_{n-1}^2(x_k)} = \frac{2}{(1-xk^2)(P_n'(x_k))^2 } $$
Задачу вычисления $ \int\limits_a^b f(x)dx $ можно свести к задаче вычисления $ \int\limits_{-1}^{1} \overline{f}(x) dx $ следующим образом
$$ \int\limits_a^b f(x)dx = [x = \frac{a+b}{2} + \frac {b-a}{2}t, dx = \frac {b-a}{2}dt] = \frac {b-a}{2} \int\limits_{-1}^{1}f(\frac {a+b}{2} + \frac {b-a}{2}t)dt = \sum\limits_{k=1}^{n}A_kf(\frac {a+b}{2} + \frac {b-a}{2}t_k) $$
Погрешность формулы НАСТ Гаусса можно вычислить следующим образом:
$$ R(f) = \frac{2^{2n+1}(n!)^4}{(2n+1)((2n)!)^3}f^{(2n)}(\xi), \xi \in [-1, 1] $$ 

## Кубатурные формулы
В общем случае необходимо вычислить
$$ \int\limits_{\Omega} f(x)dx $$
где $ x = (x_1, x_2, ... x_n), dx = dx_1dx_2...dx_n $ , $ \Omega $ - некоторая область.  
Рассмотрим случай, когда $ \Omega $ - прямоугольник.  
1 способ заключается в сведении двойного интеграла к повторному
$$ I = \int_a^b dx \int_c^d f(x,y)dy $$
$$ F(x) = \int_c^d f(x,y)dy \approx (d-c)f(x,c) $$
$$ I = \int_a^b F(x)dx \approx (a-b)(d-c)f(a,c) $$

Мы построили аналог формулы левых прямоугольников, АСТ = 0, но $ АСТ_{max} = 1 $

$ АСТ_{max} $ достигается при формуле средних прямоугольиков
$$ F(x) = \int_c^d f(x,y)dy \approx (d-c)f(x,\frac{d+c}{2}) $$
$$ I = \int_a^b F(x)dx \approx (a-b)(d-c)f(\frac{a+b}{2},\frac{d+c}{2}) $$

Аналогичным образом можно построить и формулу трапеций
$$ I = \int_a^b F(x)dx \approx \frac{S_{\Omega}}{4}(f(a,c) + f(a,d) + f(b,c) + f(b,d)) $$

2 способ заключается в использовании определения АСТ 
Строится многочлен 0 степени и определяются коэф-ты, так же делается для первой степени и вычисляем все коэф-ты, решая систему уравнений

Если облатсь произвольая то мы просто будем разбивать ее на регулярные(прямоугольники, треугольники), для которых есть хорошие формулы или строить отображения которые переведут область в регулярную, посчитать на ней, вернуться обратно

Построим формулу средних для треугольника.
$$ g(A,B,C) = 
   \begin{vmatrix}
    x_A & y_A & 1 \\
    x_B & y_B & 1 \\
    x_C & y_C & 1 \\
   \end{vmatrix}
$$
$$ \overline{x} = \frac{x_A + x_B + x_C}{3} $$
$$ \overline{y} = \frac{y_A + y_B + y_C}{3} $$
$$ I = S_{ABC} * f(\overline{x}, \overline{y}) $$

Вывдем формулу трапеций для треугольников, приблизим функцию полиномом 1 степени, т.е. лин функцию, найдем коэф-ты и по сути построим уже формулу трапеций:
$$ P_1(x,y) = \frac{1}{g(A,C,B)} [f(A)g(Q,C,B) + f(B)g(A,C,Q) + f(C)g(A,Q,B)] $$
где $ Q(x,y) $ - произвольная точка внутри треугольника
$$ I \approx \int_{\Omega}P_1(x,y)dxdy $$
Дальше можно воспользоваться формулой средних прямоугольников в итоге получим:
$$ I = \frac{1}{g(A,C,B)} [\frac{1}{2}g(A,C,B) (g(M,C,B)*f(A) + g(A,M,B)f(C) + g(A,C,M)f(B))] = \frac{S_{ABC}}{3}[f(A) + f(b) +f(c)] $$

## Интегральные уравнения Фредгольма и Вольтерра второго рода
ИУФ-2 имеет вид:
$$ u(x) - \lambda\int_{a}^{b}K(x, s)u(s)ds = f(x), x \in [a,b] , x \in R $$
$ u(x) - неизвестная функция $, $K(x,s) - ядро(задано)$  
$ \lambda - собственное значение $  
$ u - собственная функция, если (I - \lambda A)u = 0 $
### Теорема Фредгольма
$ \lambda $ не является СЗ $ \leftrightarrow $ ИУФ-2 разрешимо единственным образом  

ИУВ-2 имеет вид:
$$ u(x) - \lambda\int_{a}^{x}K(x, s)u(s)ds = f(x)  $$

### Методы решения  ИУФ-2
#### Метод квадратур
Вместо интеграла поставим квадратурную сумму.
$$ u(x) - \sum_{k=0}^{n}A_kK(x, x_k)u(x_k) = f(x) + \lambda \rho\quad(1)  $$
Пологаем $ a \leq x_0 \leq x_1 \leq ... \leq b $  
Рассмотрим случай когда в качестве квадратурной формулы используется формула $ \textbf{трапеций} $.  
Тогда $ x_k = a+kh; k=\overline{0, n} $ $ A_0 = A_n = h/2 $ все остльные $  A_k = h; k = \overline{0, n}$  
В случае когда в качестве кваратурной формулы используется формула $ \textbf{Симпсона} $.  
$ x_k = a+kh; k=\overline{0, n} $ $A_0 = A_n = h/3 $, $A_1 = A_3 ... = 4h/3 $, $ A_2 = A_4 = ... = A_{n-2} = 2h/3 $  
В каждой точке $ x_k $ запишем уравение (1) и получим систему:
$$ u(x_i) - \sum_{k=0}^{n}A_kK(x_i, x_k)u(x_k) \approx f(x_i), \quad i = \overline{0, n}  $$
Пусть $ y_k \approx  u(x_k), K_{ik} = K(x_i, x_k), f_k = f(x_k) $
Мы получили систему $Ay = f$ 
$$ A = [a_{ij}] $$
$$ a_{ii} = 1-\lambda A_{i}K_{ii} $$
$$ a_{ij} = -\lambda A_jK{ij}, \quad i \neq j $$  
Эту ситсему можно решить любым из способов.  
Решив эту систему мы получим зачения $ y_k $. Чтобы получить значение в любрй точке можно проинтерполировать полученную функцию.Либо применить след. формулу: $$ y(x) = \lambda \sum_{k=0}^{n} K(x, x_k)y_k + f(x) $$
#### Метод замены ядра на вырожденное
Ядро $ K(x, s) $ называется вырожденным, если оно представимо в виде
$$ K(x,s) = \sum\limits_{i=0}^{n} \alpha_i(x)\beta_i(s) $$
Системы $ \alpha_i(x), \beta_i(x) $ считаем линейно независимыми, так как в противном случае число слагаемых можно было бы уменьшить.  
В случае вырожденного ядра ИУФ-2 решается в аналитическом виде за конечное число действий.

Уравнение можно переписать в виде
$$ u(x) = \lambda \int\limits_a^bK(x,s)u(s)ds + f(x) = \lambda \int\limits_a^b[\sum\limits_{i=0}^{n} \alpha_i(x)\beta_i(s)]u(s)ds = $$
$$ = \lambda \sum\limits_{i=0}^{n} \alpha_i(x)\int\limits_a^b\beta_i(s)u(s)ds + f(x) = f(x) + \sum\limits_{i=0}^{n} C_i\alpha_i(x) $$
где $ C_i = \int\limits_a^b\beta_i(s)u(s)ds $
Теперь необходимо найти $ C_i $. Подставим $ u(s) $ в выражение для $ С_i $, получим
$$ C_i = \int\limits_a^b\beta_i(s)[f(s) + \sum\limits_{j=0}^{n} C_j\alpha_j(s)]ds $$
или 
$$ C_i -\lambda \sum\limits_{j=0}^{n} C_j a_{ij} = b_i$$
где $ b_i = \int\limits_a^b\beta_i(s)f(s)ds $, $ a_{ij} = \int\limits_a^b\beta_i(s)\alpha_j(s)ds $  
Остается лишь найти коэффициенты из полученной системы.

Соответсвенно можно построить приближенное решение с помощью замены ядра на вырожденное. Cамым простым способом построения такого приближения является разложение функции в ряд Тейлора. 


#### Метод последовательных приближений
Будем искать решение ИУФ-2 в виде
$$ u(x) = \sum\limits_{i=0}^{\infty}\lambda^i \phi_i(x) $$
Подставим это в уравнение и получим:
$$ \sum\limits_{i=0}^{\infty}\lambda^i \phi_i(x) - \lambda \int\limits_a^bK(x,s) [\sum\limits_{i=0}^{\infty}\lambda^i \phi_i(s)] ds = f(x) $$
Меняя порядок суммирования и интегрирования (предполагая что ряд сходится) и приравния коэф-ты при одинаковых степенях $ \lambda $ получим рекуррентные соотношения позволяющие последовательно $ \phi_i $.
$$
\begin{equation*}
    \begin{cases}
        \phi_0(x) = f(x) \\
        \phi_i(x) = \int\limits_a^b K(x,s) \phi_{i-1}(s) ds \\
    \end{cases}
\end{equation*}
$$
Введем обозначеие $ y_n(x) = \sum\limits_{i=0}^{n}\lambda^i \phi_i(x) $
Выбирая в качетсве начального приближения приближения произвольную функцию $ y_0(x) $ построим пооследовательность приближений 
$$ y_{n+1}(x) = \lambda\int\limits_a^b K(x,s)y_n(s)ds + f(x) $$
Которая при $ y_0(x) = 0 $ будет совпадать с описанным выше способом.
Рассмотрим вопрос сходимости итерационного процесса. Пусть в области $ R = [a,b] \times [a,b] $ выполняется неравенство $ |K(x,s)| \leq M $. Тогда для сходимости достаточно чтобы выполнялось 
$$ |\lambda|M(b-a) < 1 $$

### Методы решения ИУВ-2
#### Метод квадратур 
$$ u(x) - \lambda\int_{a}^{x}K(x, s)u(s)ds = f(x)  $$
Выберем на отрезке $ [a,b] $ $ (n+1) $ точку $ a \leq x_0 \leq x_1 \leq ... \leq x_n \leq b $ и рассмотрим уравнение в этих точках.
$$ u(x_i) - \lambda\int_{a}^{x_i}K(x_i, s)u(s)ds = f(x_i)  $$
Теперь в каждом из уравнений мы имеем интеграл с постоянным верхним пределом. Заменим этот интеграл квадратурной формулой использующей значение подынтегральной функции в точках $ x_0, x_1,...,x_i $
$$ u(x_i) - \lambda\sum\limits_{k=0}^{i}A_k^{(i)}K(x_i, x_k)u(x_k) = f(x_i) + \lambda\rho^{(i)} $$
Отбрасывая остаток получаем
$$ y_i - \lambda \sum\limits_{k=0}^{i}A_k^{(i)}K_{ik}y_k = f_i $$
Получили систему
$$
\begin{equation*}
    \begin{cases}
        (1-\lambda A_0^{(0)}K_{00})y_0 = f_0 \\
        -\lambda A_0^{(1)}K_{10}y_0 + (1-\lambda A_1^{(1)}K_{11})y_1 = f_1 \\
        ... \\
        -\lambda A_0^{(n)}K_{n0}y_0 - ... - \lambda A_{n-1}^{(n)}K_{n,n-1}y_{n-1} + (1-\lambda A_n^{(n)}K_{nn})y_n = f_n \\
    \end{cases}
\end{equation*}
$$

#### Метод последовательных приближений
Будем искать решение ИУВ-2 в виде
$$ u(x) = \sum\limits_{i=0}^{\infty}\lambda^i \phi_i(x) $$
Аналогично методу последовательых приближений для ИУФ-2 получаем
\begin{equation*}
    \begin{cases}
        \phi_0(x) = f(x) \\
        \phi_i(x) = \int\limits_a^x K(x,s) \phi_{i-1}(s) ds \\
    \end{cases}
\end{equation*}

В отличие от ИУФ-2 сходимость метода последовательных приближений для ИУВ-2 не накладывает ограничения на параметры исходной задачи.  

#### Метод замены ядра на вырожденное
Аналогичен методу замены ядра на вырожденное для ИУФ-2. Только коэф-ты $ C_i $ уже будут являться функциями и для их нахождения придется решать систему дифференциальных уравнений.
